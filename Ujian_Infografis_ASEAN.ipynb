#### __1. Populasi ASEAN__

Buatlah sebuah file python yang dapat mengakses daftar negara ASEAN dari database world dan menampilkan populasi negara-negara ASEAN dalam bentuk diagram batang. Output yang diharapkan berupa diagram batang dengan nilai/jumlah populasi tertera di atas diagram. Style, desain & warna tidak harus sama, utamakan validitas data:

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Extracting data from MySQL Database
dbku = mysql.connector.connect(
    host = 'localhost',
    port = 3306,
    user = 'novita',
    passwd = 'novita11',
    use_pure=True,
    database = 'world'
)
x = dbku.cursor()
x.execute('select name, population from country where region = "Southeast Asia" order by name')
df = pd.DataFrame(x.fetchall())
df.columns = ['Negara_ASEAN','Populasi_Negara']
df
```

```python
# Creating Bar Plots
plt.figure(figsize=(14, 7))
plt.style.use('seaborn')

barPlot = plt.bar(df['Negara_ASEAN'], df['Populasi_Negara'], 
              color=['royalblue','darkorange','g','firebrick','mediumpurple','sienna','plum','grey','yellowgreen','c','royalblue'],
                  width=0.5
             )

plt.title('Populasi Negara ASEAN', size=15, fontweight='bold')
plt.xlabel('Negara', size=13)
plt.xticks(rotation=45)
plt.ylabel('Populasi (x100jt jiwa)', size=13)
plt.grid(True)

def autolabel(barPlot):
    for i in barPlot:
        height = i.get_height()
        plt.text(i.get_x() + i.get_width()/2., height,
                int(height),
        ha='center', va='bottom')
autolabel(barPlot)
plt.savefig('Populasi_Negara_ASEAN.png')
plt.show()
```

#### __2. Persentase Populasi ASEAN__

Buatlah sebuah file python yang dapat mengakses daftar negara ASEAN dari database world dan menampilkan persentase populasi negara-negara ASEAN dalam bentuk diagram lingkaran. Output yang diharapkan berupa diagram lingkaran dengan nilai persentase populasi tertera dalam diagram. Style, desain & warna tidak harus sama, utamakan validitas data:

```python
plt.figure(figsize=(15, 10))
w, l, p = plt.pie(df['Populasi_Negara'], labels=df['Negara_ASEAN'], 
        colors=['royalblue','darkorange','g','firebrick','mediumpurple','sienna','plum','grey','yellowgreen','c','royalblue'],
        autopct = '%1.1f%%'
       )
plt.title('Persentase Penduduk ASEAN', size=20)
for i in p:
    i.set_color('w')
    i.set_size(12)
    i.set_fontweight('bold')

for i in l:
    i.set_size(13)
 
plt.savefig('Persentase_Penduduk_ASEAN.png')
plt.show()
```

#### __3. Gross National Product ASEAN__
Buatlah sebuah file python yang dapat mengakses daftar negara ASEAN dari database world dan menampilkan Pendapatan Bruto Nasional/GNP (Gross National Product) negara-negara ASEAN dalam bentuk diagram batang. Output yang diharapkan berupa diagram batang dengan nilai GNP tertera di atas diagram. Style, desain & warna tidak harus sama, utamakan validitas data:

```python
x.execute('select name, GNP from country where region = "Southeast Asia" order by name')
df2 = pd.DataFrame(x.fetchall())
df2.columns = ['Negara_ASEAN','GNP']
df2
```

```python
plt.figure(figsize=(14, 7))
plt.style.use('seaborn')

barPlot2 = plt.bar(df2['Negara_ASEAN'], df2['GNP'], 
              color=['royalblue','darkorange','g','firebrick','mediumpurple','sienna','plum','grey','yellowgreen','c','royalblue'],
                   width=0.5
             )

plt.title('Pendapatan Bruto Nasional ASEAN', size=20)
plt.xlabel('Negara', size=15)
plt.xticks(rotation=45)
plt.ylabel('Gross National Product (US$)', size=15)
plt.grid(True)

def autolabel(barPlot2):
    for i in barPlot2:
        height = i.get_height()
        plt.text(i.get_x() + i.get_width()/2., height,
                '%.1f' % float(height),
        ha='center', va='bottom')
autolabel(barPlot2)

plt.savefig('GNP_ASEAN.png')
plt.show()
```

#### __4. Persentase Luas Daratan ASEAN__
