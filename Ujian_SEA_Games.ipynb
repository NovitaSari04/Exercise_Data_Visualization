#### __1. Total raihan medali emas SEA Games 2017 & 2019. Berikan marker khusus pada Negara dengan raihan medali emas terbanyak.__

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DataFrame diinput manual (tanpa scraping) karena web sudah tidak bisa diakses
country = ['BRUNEI','CAMBODIA','INDONESIA','LAO PDR','MALAYSIA','MYANMAR',
           'PHILIPPINES','SINGAPORE','THAILAND','TIMOR-LESTE','VIETNAM']
gold17 = [0,3,38,2,145,7,24,57,72,0,58]
gold19 = [2,4,72,1,55,4,149,53,92,0,98]

df = pd.DataFrame(zip(country,gold17,gold19), columns=['country','2017','2019'])
df
```

```python
# Creating line plot for closing price in the last 3 months
plt.figure(figsize=(14, 7))
plt.style.use('seaborn')
plt.plot(df['country'], df['2017'], label='Gold 2017', color='r', marker='o')
plt.plot(df['country'], df['2019'], label='Gold 2019', color='g', marker='o')
plt.plot(['MALAYSIA','PHILIPPINES'],[145,149], label='Champion', marker ='*', ls='', ms='25', mec='gold',c='gold')

plt.xticks(size=11, rotation=45)
plt.yticks(np.arange(0,180,20), size=11)

plt.title('GOLD MEDAL TALLY SEA GAMES 2017 & 2019', size=19, fontweight='bold')
plt.legend(loc='upper right', fontsize=13)

plt.savefig('Medal_LinePlot.png')
plt.show()
```

#### __2. Persentase raihan medali emas SEA Games 2017 & 2019. Tampilkan nilai persentase pada diagram lingkaran.__

```python
plt.figure(figsize=(13,7))

plt.subplot(121)
w,l,p = plt.pie(df['2017'], labels=df['country'], 
        startangle=90,
        autopct='%1.1f%%',
       )
plt.title('GOLD% SEA GAMES 2017', size=20, fontweight='bold')
for i in p:
    i.set_color('w')

    
plt.subplot(122)
w,l,p = plt.pie(df['2019'], labels=df['country'],
        startangle=90,
        autopct='%1.1f%%',
       )
plt.title('GOLD% SEA GAMES 2019', size=20, fontweight='bold')
for i in p:
    i.set_color('w')

plt.savefig('Medal_PieChart.png')
plt.show()
```
