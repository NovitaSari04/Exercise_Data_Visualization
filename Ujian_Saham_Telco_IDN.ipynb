### __Soal 3 - Harga Historis Saham Provider Telco Indonesia__
Disediakan 4 buah file .csv yang berisi harga historis saham harian selama 3 bulan terakhir (18/Mar/2019 - 14/Jun/2019) dari 4 perusahaan provider telekomunikasi ternama di Indonesia.

#### __1.__
Buatlah sebuah file python yang mengakses data dari ke-4 file .csv tersebut, kemudian tampilkan data plot harga penutupan (close) harian seluruh data yang ada. Output yang diharapkan berupa grafik garis yang menampilkan harga penutupan (close) harian dari ke-4 perusahaan telco selama 3 bulan terakhir:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Extract 1st CSV file
dfXL = pd.read_csv('EXCL.JK.csv') 
```

```python
# Extract 2nd CSV file
dfFR = pd.read_csv('FREN.JK.csv')
```

```python
# Extract 3rd CSV file
dfIS = pd.read_csv('ISAT.JK.csv')
```

```python
# Extract 4th CSV file
dfTK = pd.read_csv('TLKM.JK.csv')
```

```python
# Creating line plot for closing price in the last 3 months
plt.figure(figsize=(14, 7))
plt.style.use('seaborn')
plt.plot(dfXL['Date'], dfXL['Close'], color='g', label='PT XL Axiata Tbk')
plt.plot(dfFR['Date'], dfFR['Close'], color='c', label='PT Smartfren Telecom Tbk')
plt.plot(dfIS['Date'], dfIS['Close'], color='b', label='PT Indosat Tbk')
plt.plot(dfTK['Date'], dfTK['Close'], color='r', label='PT Telekomunikasi Indonesia Tbk')

plt.xticks(dfXL['Date'][0::7], size=11, rotation=45)
plt.yticks(size=11)

plt.title('Harga Historis Saham Provider Telco Indonesia', size=19, fontweight='bold', pad=50)
plt.xlabel('Tanggal', size=15)
plt.ylabel('Rupiah (IDR)', size=15)

plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.07), ncol=4, fontsize=11)

plt.savefig('Harga_Historis.png')
plt.show()
```

#### __2.__ 
Tampilkan data plot harga penutupan (close) harian selama bulan April 2019

```python
# Creating line plot for closing price in the last 3 months
plt.figure(figsize=(14, 7))
plt.style.use('seaborn')
a = dfXL['Date'].str.contains("-04-")
b = dfFR['Date'].str.contains("-04-")
c = dfIS['Date'].str.contains("-04-")
d = dfTK['Date'].str.contains("-04-")
plt.plot(dfXL['Date'][a], dfXL['Close'][a], color='g', label='PT XL Axiata Tbk')
plt.plot(dfFR['Date'][b], dfFR['Close'][b], color='c', label='PT Smartfren Telecom Tbk')
plt.plot(dfIS['Date'][c], dfIS['Close'][c], color='b', label='PT Indosat Tbk')
plt.plot(dfTK['Date'][d], dfTK['Close'][d], color='r', label='PT Telekomunikasi Indonesia Tbk')

plt.xticks(size=11, rotation=45)
plt.yticks(size=11)

plt.title('Harga Historis Saham Provider Telco Indonesia (April 2019)', size=19, fontweight='bold', pad=50)
plt.xlabel('Tanggal', size=15)
plt.ylabel('Rupiah (IDR)', size=15)
plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.07), ncol=4, fontsize=11)

plt.savefig('Harga_Historis_April.png')
plt.show()
```
