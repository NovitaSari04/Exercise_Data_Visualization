### __Soal 2 - SPLTV (Sistem Persamaan Linear Tiga Variabel)__

Terdapat 3 buah persamaan linear 3 variabel sebagai berikut:
- x  - 2y +  z =  6
- 3x +  y - 2z =  4
- 7x - 6y -  z = 10

__1. Dengan menggunakan numpy, buatlah sebuah program python yang dapat menentukan nilai x, y & z dari 3 buah persamaan di atas!__

```python
import numpy as np
```

```python
a = np.array([[1,-2,1], [3,1,-2], [7,-6,-1]])

b = np.array([6,4,10])

x, y, z = np.linalg.solve(a,b)[0], np.linalg.solve(a,b)[1], np.linalg.solve(a,b)[2]
print('Nilai x = ',x)
print('Nilai y = ',y)
print('Nilai z = ',z)
```

__2. Dengan menggunakan matplotlib, buatlah sebuah program python yang dapat menggambarkan area 3D yang merepresentasikan masing-masing persamaan di atas. Hasil yang diharapkan saat program dieksekusi berupa 1 buah figure dengan 3 buah plot 3D__

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d 
```

```python
# Membuat plot masing-masing persamaan dalam bentuk 3D
fig = plt.figure(figsize = (15,6))

# persamaan x -2y + z = 6
x = np.array([6,0,0])   # saat y dan z = 0 maka x = 6
y = np.array([0,-3,0])  # saat x dan z = 0 maka y = -3
z = np.array([0,0,6])   # saat x dan y = 0 maka z = 6

ax = fig.add_subplot(131, projection='3d')
ax.plot_trisurf(x,y,z, color='red', alpha=0.5)
ax.scatter(x,y,z, marker='o', color='blue')
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Y')
ax.set_zlabel('Nilai Z')
plt.title('x -2y + z = 6')

# persamaan 3x + y - 2z = 4
x = np.array([4/3,0,0])   # saat y dan z = 0 maka x = 4/3
y = np.array([0,4,0])     # saat x dan z = 0 maka y = 4
z = np.array([0,0,-2])    # saat x dan y = 0 maka z = -2

ax = fig.add_subplot(132, projection='3d')
ax.plot_trisurf(x,y,z, color='olive', alpha=0.5)
ax.scatter(x,y,z, marker='o', color='red')
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Y')
ax.set_zlabel('Nilai Z')
plt.title('3x + y - 2z = 4')

# persamaan 7x - 6y - z = 10
x = np.array([10/7,0,0])    # saat y dan z = 0 maka x = 10/7
y = np.array([0,-5/3,0])    # saat x dan z = 0 maka y = -5/3
z = np.array([0,0,-10])     # saat x dan y = 0 maka z = -10

ax = fig.add_subplot(133, projection='3d')
ax.plot_trisurf(x,y,z, color='blue', alpha=0.5)
ax.scatter(x,y,z, marker='o', color='green')
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Y')
ax.set_zlabel('Nilai Z')
plt.title('7x - 6y - z = 10')

# plt.savefig('Ujian_SPLTV.png')
plt.show()
```
