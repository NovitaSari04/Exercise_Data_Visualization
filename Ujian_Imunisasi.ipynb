### __Soal 4 - Imunisasi__

Disediakan 4 buah dataset (.csv) seputar persentase balita di Indonesia yang pernah mendapatkan imunisasi BCG, Campak, DPT & Polio dalam kurun tahun 1995-2017. Buatlah sebuah program python yang dapat menampilkan data-data tersebut dalam bentuk sebagai berikut:

- Data yang missing diolah dengan metode interpolasi linear, yakni jika terdapat nilai 'n.a' akan diubah menjadi nilai antara data sebelum & sesudahnya.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca file CSV
bcg = pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv', na_values='n.a')

# Mengisi missing value dengan interpolasi data
bcg = bcg.interpolate()

# Membuat kolom % balita yang tidak mendapatkan imunisasi (bahan plot soal kedua)
bcg['% Balita yang tidak pernah mendapat imunisasi BCG'] = 100 - bcg['% Balita yang pernah mendapat imunisasi BCG']

bcg
```

```python
# Membaca file CSV
cpk = pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv', na_values='n.a')

# Mengisi missing value dengan interpolasi data
cpk = cpk.interpolate()

# Membuat kolom % balita yang tidak mendapatkan imunisasi (bahan plot soal kedua)
cpk['% Balita yang tidak pernah mendapat imunisasi Campak'] = 100 - cpk['% Balita yang pernah mendapat imunisasi Campak']

cpk
```

```python
# Membaca file CSV
dpt = pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv', na_values='n.a')

# Mengisi missing value dengan interpolasi data
dpt = dpt.interpolate()

# Membuat kolom % balita yang tidak mendapatkan imunisasi (bahan plot soal kedua)
dpt['% Balita yang tidak pernah mendapat imunisasi DPT'] = 100 - dpt['% Balita yang pernah mendapat imunisasi DPT']

dpt
```

```python
# Membaca file CSV
pol = pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv', na_values='n.a')

# Mengisi missing value dengan interpolasi data
pol = pol.interpolate()

# Membuat kolom % balita yang tidak mendapatkan imunisasi (bahan plot soal kedua)
pol['% Balita yang tidak pernah mendapat imunisasi Polio'] = 100 - pol['% Balita yang pernah mendapat imunisasi Polio']

pol
```

- Usai program dieksekusi, akan muncul 2 buah figure sekaligus. Figure 1 berisi data persentase balita yang pernah diimunisasi, dan figure 2 berisi persentase balita yang belum pernah diimunisasi.

- Setiap figure berisi 4 buah diagram batang yang menunjukkan persentase balita yang pernah diimunisasi BCG, Campak, DPT & Polio selama 1995-2017.

```python
# Figure 1: Persentasi balita terimunisasi 1995-2017
plt.figure(figsize=(11,7))

plt.subplot(221)
plt.bar(bcg['Tahun'], bcg['% Balita yang pernah mendapat imunisasi BCG'], color='r')
plt.title('BCG', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplot(222)
plt.bar(cpk['Tahun'], cpk['% Balita yang pernah mendapat imunisasi Campak'], color='g')
plt.title('Campak', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplot(223)
plt.bar(dpt['Tahun'], dpt['% Balita yang pernah mendapat imunisasi DPT'], color='y')
plt.title('DPT', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplot(224)
plt.bar(pol['Tahun'], pol['% Balita yang pernah mendapat imunisasi Polio'], color='darkblue')
plt.title('Polio', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplots_adjust(hspace = 0.3)

plt.savefig('Telah_Imunisasi.png')
plt.show()
```

```python
# Figure 1: Persentasi balita tak terimunisasi 1995-2017
plt.figure(figsize=(11,7))

plt.subplot(221)
plt.bar(bcg['Tahun'], bcg['% Balita yang tidak pernah mendapat imunisasi BCG'], color='r')
plt.title('BCG', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplot(222)
plt.bar(cpk['Tahun'], cpk['% Balita yang tidak pernah mendapat imunisasi Campak'], color='g')
plt.title('Campak', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplot(223)
plt.bar(dpt['Tahun'], dpt['% Balita yang tidak pernah mendapat imunisasi DPT'], color='y')
plt.title('DPT', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplot(224)
plt.bar(pol['Tahun'], pol['% Balita yang tidak pernah mendapat imunisasi Polio'], color='darkblue')
plt.title('Polio', size=8)
plt.xticks(np.arange(1995,2018), rotation=90, size=7)

plt.subplots_adjust(hspace = 0.3)

plt.savefig('Belum_Imunisasi.png')
plt.show()
```
